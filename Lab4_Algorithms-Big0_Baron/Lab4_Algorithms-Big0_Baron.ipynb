```python
#4. Ensuring Unique Elements with Sets 
# Step 1: Implement binary search 
def binary_search(arr, target):
    low = 0 
    high = len(arr) - 1 
    while low <= high: 
        mid = (low + high) // 2 
        if arr[mid] == target: 
            return mid 
        elif arr[mid] < target: 
            low = mid + 1 
        else: 
            high = mid - 1 
            return -1 

# Step 2: Test with a sorted list 
sorted_data = [1, 3, 5, 7, 9, 11] 
result = binary_search(sorted_data, 7) 
print("Found at index:", result) 

#Try It:  
#Trace how many comparisons are made for different target values and relate to O(log n). 


def binary_search_trace(arr, target):
    low = 0
    high = len(arr) - 1
    count = 0
    
    while low <= high:
        count += 1
        mid = (low + high) // 2
        print(f"Step {count}: Checking middle index {mid} (value: {arr[mid]})")
        
        if arr[mid] == target:
            print(f"Target found in {count} steps!")
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Running it
sorted_data = [1, 3, 5, 7, 9, 11]
binary_search_trace(sorted_data, 7)
```

```python
#5. Bubble Sort (Intro to Sorting) 

# Step 1: Implement bubble sort 
def bubble_sort(arr): 
   n = len(arr) 
   for i in range(n): 
     for j in range(0, n - i - 1): 
        if arr[j] > arr[j + 1]: 
           arr[j], arr[j + 1] = arr[j + 1], arr[j] 

# Step 2: Test sorting 
to_sort = [64, 34, 25, 12, 22, 11, 90] 
bubble_sort(to_sort) 
print("Sorted list:", to_sort) 

#Try It:  
#Count the number of swaps that occur during the bubble sort of a list.

def bubble_sort_with_swaps(arr): 
    n = len(arr) 
    swap_count = 0
    for i in range(n): 
        for j in range(0, n - i - 1): 
            if arr[j] > arr[j + 1]: 
                # A swap is happening!
                arr[j], arr[j + 1] = arr[j + 1], arr[j] 
                swap_count += 1
    
    print(f"Total swaps made: {swap_count}")

# Testing it
to_sort = [64, 34, 25, 12, 22, 11, 90] 
bubble_sort_with_swaps(to_sort) 
print("Sorted list:", to_sort)

```

```python
#6. Selection Sort 

# Step 1: Implement selection sort 
def selection_sort(arr): 
   for i in range(len(arr)): 
     min_idx = i 
     for j in range(i+1, len(arr)): 
       if arr[j] < arr[min_idx]: 
         min_idx = j 
         arr[i], arr[min_idx] = arr[min_idx], arr[i] 

# Step 2: Test and observe performance 
sample_data = [29, 10, 14, 37, 13] 
selection_sort(sample_data) 
print("Sorted list:", sample_data) 

#Try It:  
#Compare the number of comparisons and swaps in selection sort vs bubble sort. 

def selection_sort_stats(arr):
    comparisons = 0
    swaps = 0
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            comparisons += 1  # Every 'if' is a comparison
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        # We swap once per outer loop
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        swaps += 1
        
    print(f"Selection Sort -> Comparisons: {comparisons}, Swaps: {swaps}")

# Testing with [29, 10, 14, 37, 13]
sample_data = [29, 10, 14, 37, 13]
selection_sort_stats(sample_data)

```

```python
# 7. Insertion Sort
 
# Step 1: Implement insertion sort 
def insertion_sort(arr): 
    for i in range(1, len(arr)):
        key = arr[i] 
        j = i - 1 
    while j >= 0 and key < arr[j]: 
            arr[j + 1] = arr[j] 
            j -= 1 
            arr[j + 1] = key 

# Step 2: Test and measure performance 
arr = [12, 11, 13, 5, 6] 
insertion_sort(arr) 
print("Sorted array:", arr) 

#Try It:  
#Analyze insertion sort’s performance when the list is already sorted. 

def insertion_sort_trace(arr):
    comparisons = 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        # This condition is the 'Check'
        while j >= 0:
            comparisons += 1
            if key < arr[j]:
                arr[j + 1] = arr[j]
                j -= 1
            else:
                break # STOP! The card is already in the right spot.
        
        arr[j + 1] = key
    print(f"Total comparisons: {comparisons}")

# Test with a sorted list
sorted_list = [5, 6, 11, 12, 13]
insertion_sort_trace(sorted_list)
```

```python
# 3. Sorting Algorithm Visual Comparator

import time
import random

# --- INPUTS ---
student_id = 253829
list_size = int(str(student_id)[-3:]) + 50
data = random.sample(range(10000), list_size)

def bubble_sort(arr):
    n, comps, swaps = len(arr), 0, 0
    for i in range(n):
        for j in range(0, n - i - 1):
            comps += 1
            if arr[j] > arr[j + 1]:
                swaps += 1
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return comps, swaps

def selection_sort(arr):
    comps, swaps = 0, 0
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            comps += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        swaps += 1
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return comps, swaps

def insertion_sort(arr):
    comps, swaps = 0, 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            comps += 1
            swaps += 1
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return comps, swaps

# --- EXECUTION & OUTPUT ---
print(f"{'Algorithm':<15} {'Comps':<10} {'Swaps':<10} {'Time (sec)'}")
print("-" * 50)

for name, func in [("Bubble", bubble_sort), ("Selection", selection_sort), ("Insertion", insertion_sort)]:
    start = time.time()
    c, s = func(data.copy())
    t = time.time() - start
    print(f"{name:<15} {c:<10} {s:<10} {t:.6f}")

"""
ANALYSIS & RECOMMENDATION:
While all three algorithms are O(n^2), Insertion Sort is generally the 
most efficient on small or random datasets because it performs fewer 
unnecessary swaps. Bubble Sort is typically the slowest due to the high 
overhead of repeated adjacent swaps. Recommendation: For better efficiency 
on larger datasets, one should move to O(n log n) algorithms like Merge Sort 
or Timsort (Python's built-in .sort()).
"""
```
